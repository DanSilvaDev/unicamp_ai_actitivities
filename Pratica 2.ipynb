# Importar as bibliotecas e carregar a LLM

````python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "mistralai/Mistral-7B-Instruct-v0.3"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
````

# Entendendo o Tokenizador

````python
prompt = "Diga as qualiades do LLM mistral."
inputs = tokenizer(prompt)
# imprime os tokens de entrada
print(inputs)
# converte os tokens de entrada em um tensor
inputs_tensor = tokenizer(prompt, return_tensors="pt")
# imprime o tensor de entrada
print(inputs_tensor)
````

# Criando um Pipeline

````python
from transformers import pipeline
# Create a pipeline
generator = pipeline(
"text-generation",
model=model,
tokenizer=tokenizer,
return_full_text=False,
max_new_tokens=500,
do_sample=False
)
````

# Gerando uma resposta

````python
# Generate output
output = generator(prompt)
print(output[0]["generated_text"])
````
